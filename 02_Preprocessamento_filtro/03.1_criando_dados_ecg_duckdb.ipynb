```python
import wfdb
import duckdb
import time
import pandas as pd

# Iniciar contagem do tempo
tempo_inicio = time.time()

# # Conectar ao banco de dados DuckDB
conn = duckdb.connect("D:/Projeto_Tese_mestrado/02_Dataset/Duckedb/Holter_ECG/1h/banco_ecg.duckdb")

# Criar a tabela se n√£o existir
conn.execute("""
    CREATE TABLE IF NOT EXISTS ecg_pacientes (
        id_paciente TEXT,
        amostra INTEGER,
        sinal_x FLOAT,
        sinal_y FLOAT,
        sinal_z FLOAT
    )
""")

# Par√¢metros de segmenta√ß√£o
inicio_amostra = 720000  
duracao_em_amostras = 720000
# Definir o tamanho do lote de pacientes
tamanho_lote = 200

# Loop sobre os arquivos sem duplicatas
for idx, endereco in enumerate(lista_sem_duplicatas):
    try:
        print(f"üîÑ Processando paciente {idx+1}/{len(lista_sem_duplicatas)}: {endereco}")

        # Lendo o sinal do ECG
        record = wfdb.rdrecord(endereco)
        p_signal = record.p_signal

        # Verificar se h√° pelo menos 3 canais
        if p_signal.shape[1] < 3:
            print(f"‚ö†Ô∏è Paciente {record.record_name} tem menos de 3 canais de ECG. Pulando...")
            continue

        # Obter sinais X, Y e Z
        sinal_x = p_signal[:, 0]
        sinal_y = p_signal[:, 1]
        sinal_z = p_signal[:, 2]

        # Filtrar os sinais
        sinal_fil_x = sinal_x[inicio_amostra: inicio_amostra + duracao_em_amostras]
        sinal_fil_y = sinal_y[inicio_amostra: inicio_amostra + duracao_em_amostras]
        sinal_fil_z = sinal_z[inicio_amostra: inicio_amostra + duracao_em_amostras]

        # Identifica√ß√£o do paciente
        id_paciente = record.record_name  

        # # Criar uma lista com os dados
        # dados_geral = []
        # for i in range(len(sinal_fil_x)):
        #     dados_geral.append([id_paciente, i, sinal_fil_x[i], sinal_fil_y[i], sinal_fil_z[i]])

        # # Criar um DataFrame com os dados do paciente
        # df_ecg = pd.DataFrame(dados_geral, columns=["id_paciente", "amostra", "sinal_x", "sinal_y", "sinal_z"])
        df_ecg = pd.DataFrame({
            "id_paciente": id_paciente,
            "amostra": range(len(sinal_fil_x)),
            "sinal_x": sinal_fil_x,
            "sinal_y": sinal_fil_y,
            "sinal_z": sinal_fil_z
        })


        # Inserir os dados no banco em lote
        if not df_ecg.empty:
            print(f"üìÇ Inserindo dados do paciente {id_paciente} no banco...")
            #conn.execute("INSERT INTO ecg_pacientes SELECT * FROM df_ecg")
            conn.execute("BEGIN")
            conn.execute("INSERT INTO ecg_pacientes SELECT * FROM df_ecg")
            conn.execute("COMMIT")

            print(f"‚úÖ Dados do paciente {id_paciente} inseridos com sucesso!")

        # Liberar mem√≥ria ap√≥s cada lote
        del df_ecg
        #dados_geral.clear()

        # A cada 200 pacientes, limpar a mem√≥ria e inserir novos dados
        if (idx + 1) % tamanho_lote == 0:
            print(f"üíæ Lote de {tamanho_lote} pacientes processado. Aguardando pr√≥ximo lote...")

    except Exception as e:
        print(f"‚ùå Erro ao processar {endereco}: {e}")
        continue  # Continua para o pr√≥ximo paciente

# Fechar conex√£o com o banco
conn.close()

# Finalizar contagem do tempo
tempo_fim = time.time()
tempo_total = tempo_fim - tempo_inicio

# Exibir o tempo total de execu√ß√£o
print(f"‚è≥ Tempo total de execu√ß√£o: {tempo_total:.2f} segundos")


```

```python
# contando a quantidade por pacientes
import duckdb
import time
import pandas as pd

# Iniciar contagem do tempo
tempo_inicio = time.time()

# Conectar ao banco DuckDB
conn = duckdb.connect("D:/Projeto_Tese_mestrado/02_Dataset/Duckedb/Holter_ECG/1h/banco_ecg.duckdb")

# Consulta: contar amostras por paciente
df = conn.execute("""
    SELECT id_paciente, COUNT(*) AS total_amostras
    FROM ecg_pacientes
    GROUP BY id_paciente
    ORDER BY id_paciente
""").fetchdf()

# Exibir resultado
print(df)

# Fechar conex√£o
conn.close()

# Tempo de execu√ß√£o
tempo_fim = time.time()
print(f"‚è±Ô∏è Tempo de execu√ß√£o: {tempo_fim - tempo_inicio:.2f} segundos")
```
