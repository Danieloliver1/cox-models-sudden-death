```python
import neurokit2 as nk
import pandas as pd
import numpy as np

# 1. Carrega um sinal de ECG (aqui é simulado, mas pode ser de um arquivo)
ecg_signal = nk.ecg_simulate(duration=3000, sampling_rate=500)  # 10s de ECG sintético


```

```python
import neurokit2 as nk

# 1. ECG cru (simulado ou real)

sampling_rate = 500

# 2. Processa o sinal com o NeuroKit
ecg_processed, info = nk.ecg_process(ecg_signal, sampling_rate=sampling_rate)



```

```python

# 3. Cria épocas com base no sinal processado (e não o sinal cru!)
# Exemplo: duas épocas de 30 segundos (0 a 30s e 30s a 60s)
epochs = nk.epochs_create(ecg_processed,
                          events=[0, 15000],  # índices dos eventos de início
                          sampling_rate=sampling_rate,
                          epochs_start=0,
                          epochs_end=30)  # segundos
```

```python
# 4. Extrai características estatísticas por época
features = nk.ecg_intervalrelated(epochs)

```
